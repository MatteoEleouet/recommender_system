# Système de recommandation avec Python
### Premier système de recommandation

```python
import numpy as np
import pandas as pd
```

### Récupération des données
    Nous avons deux jeux de données :
        Une colonne du nombre d'avis
        Une colonne des films avec leur nom et leur ID

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('movies.csv', sep='\t', names=column_names)
```

```python
df.head(10)
```

```python
movie_titles = pd.read_csv("movie_id_titles.csv")
movie_titles.head(10)
```

```python
df = pd.merge(df,movie_titles,on='item_id')
df.head(10)
```

```python
df = df.drop(columns=['title_y'], axis=1)
df = df.rename(columns={"title_x" : 'title'})
df.head()
```

## Analyse exploratoire
    Explorons un peu les données et jetons un coup d'œil à certains des films les mieux notés.

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
%matplotlib inline
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=True).head()
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['num of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
plt.figure(figsize=(14,6))
ratings['num of ratings'].hist(bins=40, color='navy', alpha=0.9, label="Quantité d'évaluation") # je me suis amusé avec Matplotlib
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,4))
ratings['rating'].hist(bins=40)
```

Globalement les évalutions se trouve entre 2 et 4

## Recommandation de film par similitude

```python
moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

Il y a énormément de valeurs NaN car les gens n'ont pas vu la plupart des films présent sur la plateforme

```python
ratings.sort_values('num of ratings',ascending=False).head(10)
# Les films ayant reçu le plus de note
```

Je choisis 2 films qu'on va analyser pour recommender aux utilisateurs un film selon un film

```python
ratings.head()
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
dumb_user_ratings = moviemat['Liar Liar (1997)']
starwars_user_ratings.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similar_to_liarliar = moviemat.corrwith(dumb_user_ratings)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_starwars = corr_starwars.join(ratings['num of ratings'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```

On fait de même pour le film Liar Liar

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar,columns=['Correlation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar = corr_liarliar.join(ratings['num of ratings'])
corr_liarliar[corr_liarliar['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```

C'est terminé !

Les résultats sont assez logiques, si l'utilisateurs regard le premier star wars, on lui propose d'autre Star Wars.
